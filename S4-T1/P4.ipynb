```python
q1 = df['4th'].quantile(0.25)
q3 = df['4th'].quantile(0.75)
iqr = q3 - q1
lr = q1 - 1.5 * iqr
ur = q3 + 1.5 * iqr
out = df[(df['4th'] < lr) | (df['4th'] > ur)]
print(out)
```
