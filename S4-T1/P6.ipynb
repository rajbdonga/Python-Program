# Folium

```python
import pandas as pd
import folium
df = pd.read_csv("Police_Department_Incidents_-_Previous_Year__2016_.csv")
df.head()
```

```python
df.shape
```

```python
sanfran = folium.Map(location=[37.22 , -122.42],zoom_start=5)
sanfran
```

```python
limit = 100
df = df.iloc[0:limit , :]
df.shape
```

```python
incidents = folium.map.FeatureGroup()
for lat , lon in zip(df.Y , df.X):
    incidents.add_child(folium.features.CircleMarker([lat , lon] , radius = 5 , color = 'yellow' ,
                                                    fill = True , fill_color = 'blue' , fill_opacity = 0.6))
sanfran.add_child(incidents)
```

# Network

```python
import networkx as nx
```

```python
import matplotlib.pyplot as plt
```

```python
g = nx.Graph()
g.add_node(1)
g.add_nodes_from([2,3])
g.add_nodes_from(range(4,7))
g.add_edge(1,2)
g.add_edge(1,1)
g.add_edges_from([(2,3),(3,6),(4,6),(5,6)])
nx.draw_networkx(g , node_size = 600 , node_color = 'red' , width = 5 , edge_color = 'blue')
plt.show()
```

```python
g = nx.DiGraph()
g.add_node(1)
g.add_nodes_from([2,3])
g.add_nodes_from(range(4,7))
g.add_edge(1,2)
g.add_edge(1,1)
g.add_edges_from([(2,3),(3,6),(4,6),(5,6)])
nx.draw_networkx(g , node_size = 600 , node_color = 'red' , width = 5 , edge_color = 'blue')
plt.show()
```

# Adjacency Matrix

```python
g = nx.DiGraph()
l = [[1,0,0,1,0,1,1],[0,1,1,1,1,1,0],[1,1,1,0,1,0,1]]
g.add_nodes_from([0,1,2])
for i in range(3):
    for j in range(3):
        if l[i][j] == 1:
            g.add_edge(i,j)
nx.draw_networkx(g)
```
