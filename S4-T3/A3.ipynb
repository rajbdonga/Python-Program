```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_state_and_union_territory_capitals_in_India'
```

```python
response = requests.get(url)
response
```

```python
soup = BeautifulSoup(response.text,'html.parser')
```

```python
tables = soup.find_all('table',{'class':'wikitable'})
```

```python
tables
```

###### Extract table data

```python
def extract_table(table):
    headers = [header.get_text(strip = True) for header in table.find_all('th')]
    
    # Extract row
    rows = table.find_all('tr')[1:]
    
    # List of table data
    tabledata = []
    
    # Iterate through rows
    for row in rows:
        cells = row.find_all('td')
        rowdata = [cell.get_text(strip = True) for cell in cells]
        tabledata.append(rowdata)
        
    return headers,tabledata
```

```python
headers1,data1 = extract_table(tables[0])
data1
```

```python
headers2,data2 = extract_table(tables[1])
```

```python
df_states = pd.DataFrame(data1,columns = headers1)
df_union = pd.DataFrame(data2,columns = headers2)
```
