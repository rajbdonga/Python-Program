```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
with open("140 Essential Action Movies To Watch Now ｜ Rotten Tomatoes (5_28_2025 8：24：58 AM).html",'r',encoding='utf-8') as f:
    content = f.read()
    soup = BeautifulSoup(content,'html.parser')
```

```python
def gettage(headings):
    titles = []
    years = []
    scores = []
    for heading in headings:
        title = heading.find('a').text.strip()
        year = heading.find('span',class_ = 'start-year').text.strip()
        score = heading.find('span',class_ = 'tMeterScore').text.strip()
        titles.append(title)
        years.append(year)
        scores.append(score)
    return titles,years,scores
```

```python
divs = soup.find_all('div',class_ = 'col-sm-18 col-full-xs countdown-item-content')
headings = [div.find('h2') for div in divs]
```

```python
tags = gettage(headings)
```

```python
moviedict = {
    'title' : tags[0],
    'year' : tags[1],
    'score' : tags[2]
}
```

```python
df = pd.DataFrame(moviedict)
df
```

```python
for i in df['year']:
    df['year'] = i.strip('()')
    print(df)
```
