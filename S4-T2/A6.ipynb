###### 1. import libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix,accuracy_score
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df
```

###### 2. Preprocessing

```python
df['mathpass'] = np.where(df['math score'] >= 50,1,0)
df['readpass'] = np.where(df['reading score'] >= 50,1,0)
df['writepass'] = np.where(df['writing score'] >= 50,1,0)
df
```

###### 3. Drop Colummns

```python
df.drop(['math score' , 'reading score' , 'writing score'] , axis = 1 , inplace = True)
```

```python
df
```

###### 4. Create Overall Pass Column

```python
df['overallpass'] = np.where(
(df['mathpass'] == 1) &
(df['readpass'] == 1) &
(df['writepass'] == 1),1,0
)
```

```python
df
```

###### 5. One Hot Encoding (get dummies)

```python
df.info()
```

```python
category = ['gender','race/ethnicity','parental level of education','lunch','test preparation course']
df_encode = pd.get_dummies(df.drop(columns = 'mathpass'),columns = category , drop_first = True)
```

```python
df_encode
```

###### 6. Handel Missing Value

```python
df_encode.isna().sum()
```

###### 7. Extract Fetures and Label

```python
x = df_encode
y = df['mathpass']
```

###### 8. Split the Data

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2 , random_state = 10)
```

```python
print(x.shape)
print(x_train.shape)
print(x_test.shape)
print(y.shape)
print(y_train.shape)
print(y_test.shape)
```

###### 9. KNN CLASSIFICATION

```python
bestaccuracy = 0
bestk = 1
for k in range(1,20):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train,y_train)
    y_pred = knn.predict(x_test)
    accuracy = accuracy_score(y_test,y_pred)
    if accuracy > bestaccuracy:
        bestk = k
        bestaccuracy = accuracy
print("Best accuracy is " , bestaccuracy)
print("Best k is " , bestk)
```

###### 10. Train with best knn

```python
knn = KNeighborsClassifier(n_neighbors=bestk)
knn.fit(x_train,y_train)
y_pred_knn = knn.predict(x_test)
```

###### 11. Confusion Matrix of knn

```python
cm_knn = confusion_matrix(y_test , y_pred_knn)
print(cm_knn)
```

```python
tn , fp , fn , tp = cm_knn.ravel()
```

```python
Sensitivity = tp/(tp+fn)
Specificity = tn/(tn+fp)
accuracy_knn = accuracy_score(y_test,y_pred_knn)
print(Sensitivity)
print(Specificity)
print(accuracy_knn)
```

###### 12. Decision Tree

```python
bestdepth = 1
bestacctree = 0
for depth in range(1,11):
    tree = DecisionTreeClassifier(max_depth=depth,random_state=10)
    tree.fit(x_train,y_train)
    y_pred = tree.predict(x_test)
    accuracy = accuracy_score(y_test,y_pred)
    if accuracy > bestacctree:
        bestdepth = depth
        bestacctree = accuracy
print(bestdepth,bestacctree)
```

###### 13. fit decision tree with best depth

```python
tree = DecisionTreeClassifier(max_depth=bestdepth,random_state=10)
tree.fit(x_train,y_train)
```
